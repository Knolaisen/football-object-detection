```python
from ultralytics import YOLO
```

```python
model = YOLO("yolo11n.pt")  # load a custom model
data_path = "data/k_fold/0/data.yaml"  # path to your data.yaml file
```

```python
model.train(
    data=data_path,  # path to your data.yaml file
    epochs=1,  # number of training epochs
    batch=16,  # batch size
    imgsz = 1280,
    device="0",  # device to train on (0 for GPU, 'cpu' for CPU)
    workers=4,  # number of workers for data loading
    project="runs/train",  # directory to save the results
    name="baseline",  # name of the experiment
    exist_ok=True,  # overwrite existing experiment
)
```

```python
results = model("test_images/21.jpg")
```
